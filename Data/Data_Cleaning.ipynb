## Final Dataset merging

Here all the datasets are merged into a singular dataframe and a singular csv

```python
import pandas as pd
import numpy as np
import os

df_crypto = pd.read_csv('coinmetrics_crypto.csv')

df_crypto = df_crypto.rename(columns={'time': 'date', 'asset' : 'symbol'})
df_crypto['symbol'] = df_crypto['symbol'].str.upper()
df_crypto['date'] = pd.to_datetime(df_crypto['date'])
df_crypto['date'] = df_crypto['date'].dt.tz_localize(None)
df_crypto = df_crypto.dropna(subset='PriceUSD')
df_crypto.info()
```

```python
import pandas as pd

df_gold = pd.read_csv('Cleaned Dataset/new_gold_data.csv')

df_gold['date'] = pd.to_datetime(df_gold['date'])
df_gold = df_gold.rename(columns={'close': 'PriceUSD'})
df_gold = df_gold.drop(columns=['high', 'low', 'open', 'close'], errors='ignore')

df_gold.info()
```

```python
df_final = pd.concat([df_crypto, df_gold], ignore_index=True)

df_final
```

```python
df_feg = pd.read_csv('Cleaned Dataset/new_fear_and_greed.csv')
df_feg['date'] = pd.to_datetime(df_feg['date'])
df_feg.drop(columns=['Unnamed: 0'], inplace=True, errors='ignore')

df_vix = pd.read_csv('Cleaned Dataset/new_vix.csv')
df_vix['date'] = pd.to_datetime(df_vix['date'])

df_cosent = pd.read_csv('Cleaned Dataset/new_consumer_sentiment.csv')
df_cosent['date'] = pd.to_datetime(df_cosent['date'])

df_twsent = pd.read_csv('Cleaned Dataset/new_twitter_sentiment.csv')
df_twsent['date'] = pd.to_datetime(df_twsent['date'])

df_gem = pd.read_csv('Cleaned Dataset/new_GEM.csv')
df_gem['date'] = pd.to_datetime(df_gem['date'])

df_epu = pd.read_csv('Cleaned Dataset/new_EPU.csv')
df_epu['date'] = pd.to_datetime(df_epu['date'])
df_epu.drop(columns=['level_0_x', 'index_x', 'level_0_y', 'index_y'], inplace=True, errors='ignore')

df_insent = pd.read_csv('Cleaned Dataset/new_investor_sentiment.csv')
df_insent['date'] = pd.to_datetime(df_insent['date'])

```

```python
from functools import reduce

macro_dfs = [df_feg, df_vix, df_cosent, df_twsent, df_gem, df_epu, df_insent]
for df in macro_dfs:
    df['date'] = pd.to_datetime(df['date'])

df_final['date'] = pd.to_datetime(df_final['date'])

df_final = reduce(lambda left, right: pd.merge(left, right, on='date', how='left'), [df_final] + macro_dfs)
df_final
```

```python
# These lists are used to categorize the cryptocurrencies

altcoin_list = ['1INCH', 'AAVE', 'ADA', 'ALCX', 'ALGO', 'ALPHA', 'ANT', 'API3', 'ATOM',
                'AUDIO', 'AVAX', 'BADGER', 'BAL', 'BAT', 'BCH', 'BIT', 'BNB', 'BNT',
                'BSV', 'CRO', 'CEL', 'COMP', 'CRV', 'CVC', 'CVX', 'DASH', 'DCR', 'DGB',
                'DOT', 'DRGN', 'ELF', 'ENJ', 'ENS', 'EOS', 'ETC', 'ETH', 'FIDA', 'FIL',
                'FLOW', 'FTT', 'FUN', 'FXS', 'GALA', 'GAS', 'GLM', 'GNO', 'GNT', 'GRIN',
                'GRT', 'HEDG', 'HT', 'ICP', 'ICX', 'KNC', 'LDO', 'LEND', 'LINK', 'LOOM',
                'LPT', 'LRC', 'LSK', 'LTC', 'LUNA', 'MAID', 'MANA', 'MKR', 'MTL_METAL',
                'NEAR', 'NEO', 'NMR', 'OGN', 'OMG', 'PAY', 'PERP', 'POLY', 'POWR', 'PPT',
                'QASH', 'QNT', 'QTUM', 'REN', 'REP', 'ROOK', 'RSR', 'SAND', 'SKL', 'SNT',
                'SNX', 'SOL', 'SRM', 'STORJ', 'SUSHI', 'SWRV', 'TRX', 'UMA', 'UNI', 'VET',
                'VTC', 'WAVES', 'WNXM', 'WTC', 'XEM', 'XLM', 'XMR', 'XRP', 'XTZ',
                'XVG', 'YFI', 'ZEC', 'ZIL', 'ZRX','BTG']
other = ['WBTC', 'RENBTC', 'HBTC', 'WETH']
meme_list = ['SHIB', 'DOGE', 'APE']
goldpeg_list = ['XAUT', 'PAXG']
stable_list = ['USDT', 'USDC', 'BUSD', 'DAI', 'GUSD', 'HUSD', 'PAX', 'TUSD']

df_final['is_alt'] = df_final['symbol'].isin(altcoin_list).astype(int)
df_final['is_meme'] = df_final['symbol'].isin(meme_list).astype(int)
df_final['is_gp'] = df_final['symbol'].isin(goldpeg_list).astype(int)
df_final['is_stable'] = df_final['symbol'].isin(stable_list).astype(int)
df_final = df_final[~df_final['symbol'].isin(other)]
df_final['is_btc'] = df_final['symbol'].isin(['BTC']).astype(int)
df_final['is_gold'] = df_final['symbol'].isin(['GLDBLN']).astype(int)

df_final = df_final[(df_final['date'] <= '2025-05-31') & (df_final['date'] >= '2015-01-01')]

df_final = df_final.sort_values(by=['symbol', 'date']).copy()
# df_final.loc[:, 'daily_return'] = (df_final['close'] - df_final['open']) / df_final['open']
df_final.loc[:, 'daily_return'] = df_final.groupby('symbol')['PriceUSD'].pct_change(fill_method=None)


df_final.to_csv('new_cleaned_crypto_data.csv', index=True)
```
